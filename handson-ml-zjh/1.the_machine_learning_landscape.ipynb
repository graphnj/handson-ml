# The Machine Learning Landscape

**机器学习**是*learn from data*的科学

系统用于学习的已知数据成为训练集training set，其中的单个训练数据称为training instance(Sample)

机器学习的类型：
1. 是否需要人工监督（监督学习supervised，无监督学习unsupervised，半监督学习semisupervised，强化学习Reinforcement Learning）
2. 是否可在线对增量数据学习（在线学习online vs. 批量学习batch learning）
3. 是简单讲新数据与已知数据进行比较，还是取而代之的通过分析训练数据的模式并得到一个预测模型，可分为（基于实例的instance-based vs. 基于模型的 model-based）

## Supervised/Unsupervised Learning
### supervised learning
训练数据包含标签labels，有的资料中也称为target。
典型的监督学习主要包括分类classification与回归regression。分类算法，如垃圾邮件分类系统，需要预先对训练数据在labels中指定其分类，可以是二分类也可是多分类，回归则用于数值的预测。

机器学习中attribute指数据项，feature根据上下文可能有很多意思，通常表示attribute及其取值，如age是一个attribute，age=20则是一个feature。但大部分人这两个词都是混用的。

1. k-Nearest Neighbors
2. Linear Regression
3. Logistic Regression
4. Support Vector Machines(SVMs)
5. Decision Trees and Random Forests
6. Neural networks(某些神经网络算法如自编码及受限Boltzmann机为非监督的，刺猬也有半监督的如deep belief network。。。)

### unsupervised learning
无监督学习的训练数据没有对应label字段
常用无监督学习算法：
1. Clustering（K-Means，Hierarchical Cluster Analysis，EM）
2. Visualization & dimensionality reduction（PCA，KernelPCA，Locally-Linear Embedding）
3. Association rule learning(Apriori,Eclat)

### semisupervised learning
训练数据中只有部分有标签，通常大部分数据unlabeld，一小部分有标签。
大部分半监督学习由无监督学习和监督学习算法组合而来，如deep belief network(DBNs)基于restricted boltzmann machines(RBMs),后者RBM是个无监督的训练过程，但最终整个系统又是监督学习技术来调优的。

### reinforcement learning
强化学习系统中，有一个agent的角色可以观察整体系统，可以做出选择及执行动作，并可对其行为做出奖惩措施，于是agent可以学习一个最佳策略，使得系统的收益最大化。 ...机器人行走...

## 批量Batch及在线学习online
批量学习系统中系统不能增量数据中国学习，必须采用所有可用数据去训练，这将耗费很大的时间及计算资源。一旦系统训练完成并部署到生产环境，便不能再学习了，仅仅是应用已经学习到的知识。也叫离线学习  offline learning。
如果要用新数据，必须重新训练新版本..

批量学习的训练，评估及上线过程可以自动化进行，但是其训练时间长，数据量一旦大了后只能按天或按周的训练。总之，批量学习就是，慢，贵，性价比低，有时甚至无法实现。

幸好我们可以选择Online在线学习方式。

在online learning中，可以连续顺序的将单个数据或小批量数据送入系统，用于增量训练过程，每个步骤都性价比高。因而可以在运行中学习到新数据。

online learning对连续接收数据的系统及需要对学习的结果能快速应用很有用。另一方面，如果计算资源有限，也可采用online learning，将已经处理过的实时数据直接丢弃。对于巨大的存量数据集则可分而治之依次处理。

注：整个online learning的学习过程通常是线下进行的，名称有点confuse，可以理解为增量学习，incremental learning。

online learning系统中一个参数是间隔多久采用新数据训练，也称为learning rate。对高学习率，则学得快忘得快，低学习率则学得慢，但同时忘得慢。


## 主要挑战
bad algorithm、bad data

1. 训练数据不足
2. 无表征意义的训练数据及样本偏差
3. 低质量的训练数据（错误，离群点及噪音数据）
4. 不相关的特征，需要做特征工程，如特征选择，特征提取,必要时候要构建新特征
5. 过拟合（模型太复杂，训练数据不足，将异常数据都学习了）--简化模型，收集更多数据，修复数据错误
6. 欠拟合
